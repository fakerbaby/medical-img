```python
import os
import numpy
import pandas as pd
import shutil
import random
import re
```

```python
path = "../dataset/train/train_AB_light/"
list = os.listdir(path)
```

```python
list_A_P, list_A_N, list_B_P, list_B_N = [], [], [], []
for l in list:
    if not re.match("A.*P0.*", l):
        list_A_P.append(l)
    elif re.match("A.*P0.*", l):
        list_A_N.append(l)
    elif  re.match("B.*P0.*", l):
        list_B_N.append(l)
    elif not re.match("B.*P0.*", l):
        list_B_P.append(l)
```

```python
def move_file(obj, target, custom_list=None, r=None):
    if list == None:
        pahtdir = os.listdir(obj)
    else:
        pahtdir = custom_list
    file_len = len(pahtdir)
    rate = r if r is not None else 0.5
    pickname = int(file_len * rate)
    sample = random.sample(pahtdir, pickname)
    print(obj)
    for name in sample:
        print(name)
        shutil.copy(os.path.join(obj, name), os.path.join(target, name))
    return sample
```
