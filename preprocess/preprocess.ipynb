```python
"""
split dataset to T and F
"""
def split_dataset_to_T_and_F(src, target):
    origin_list = os.listdir(src)
    trainset_list_T, trainset_list_F = [], []
    for file in origin_list:
        if file.endswith('T.bmp'):
            shutil.copy(os.path.join(src, file), os.path.join(target[0], file))
            print("T:",file)
            trainset_list_T.append(file)
        elif file.endswith('N.bmp'):
            shutil.copy(os.path.join(src, file), os.path.join(target[1], file))
            print("N:",file)
            trainset_list_F.append(file)
    return trainset_list_T, trainset_list_F
    
```
